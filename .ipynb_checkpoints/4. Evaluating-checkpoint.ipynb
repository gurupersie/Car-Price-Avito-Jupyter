```python
df = pd.read_csv('./data_store/cleaned_data.csv')
cols_to_convert = [
    col for col in df.select_dtypes('float64').columns if col != 'Рейтинг'
]
df[cols_to_convert] = df[cols_to_convert].astype('int64')
df['Цена'] = df['Цена'].astype('int32')
df['Год_выпуска'] = df['Год_выпуска'].astype('int16')
df['Пробег_км'] = df['Пробег_км'].astype('int32')
df['Мощность_двигателя_лс'] = df['Мощность_двигателя_лс'].astype('int16')
df['Объём_двигателя_л'] = df['Объём_двигателя_л'].astype('float32')
df['Рейтинг'] = df['Рейтинг'].astype('float32')
df[df.select_dtypes('int64').columns] = df[df.select_dtypes(
    'int64').columns].astype('int8')
df[df.select_dtypes('object').columns] = df[df.select_dtypes(
    'object').columns].astype('category')
```

```python
df.info()
```

```python
X = df.drop('Цена', axis=1)
y = df['Цена']
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.15,
                                                    shuffle=True,
                                                    random_state=SEED)
X_train_, X_val, y_train_, y_val = train_test_split(X_train,
                                                    y_train,
                                                    test_size=0.1,
                                                    shuffle=True,
                                                    random_state=SEED)
cat_feature_names = list(X_train.select_dtypes('category').columns)
```

# Evaluating and Prediction intervals.

Дополнительно к оценке самих предсказаний, также предскажем интервал, в пределах которого с вероятностью 95% находится наше истинное значение. Начнём с оценок предсказания средних.

```python
tuned_catb = CatBoostRegressor(**CATB_PARAMS)
tuned_lgbm = LGBMRegressor(**LGBM_PARAMS)
tuned_catb.fit(X_train_,
               y_train_,
               eval_set=[(X_val, y_val)],
               early_stopping_rounds=50)
tuned_lgbm.fit(X_train_,
               y_train_,
               eval_set=[(X_val, y_val)],
               eval_metric='mae')
catb_preds = tuned_catb.predict(X_test)
mean_catb_preds, var_catb_preds = catb_preds[:, 0], catb_preds[:, 1]
lgbm_preds = tuned_lgbm.predict(X_test)
```

Экспериментально выяснено - показатели в пределах 20% нормальны. Catboost переобучился, LGBM в пределах нормы.

```python
metrics= pd.DataFrame()
metrics = concat_metrics(
    metrics,
    y_train.values,
    tuned_catb.predict(X_train)[:, 0],
    X_train,
    model_name=f'{tuned_catb.__class__.__name__}_tuned_train')
metrics = concat_metrics(
    metrics,
    y_test.values,
    mean_catb_preds,
    X_test,
    model_name=f'{tuned_catb.__class__.__name__}_tuned_test')
metrics = concat_metrics(
    metrics,
    y_train.values,
    tuned_lgbm.predict(X_train),
    X_train,
    model_name=f'{tuned_lgbm.__class__.__name__}_tuned_train')
metrics = concat_metrics(
    metrics,
    y_test.values,
    lgbm_preds,
    X_test,
    model_name=f'{tuned_lgbm.__class__.__name__}_tuned_test')
metrics
```

Посмотрим интервал для Catboost. Используем специальный лосс для этих целей - RMSEWithUncertainty, который мы подавали в качестве гиперпараметра внутри objective. Он позволит нам получать двухмерные предсказания: среднего и дисперсии. Выберем коэффициент 1.96 - он соответствует 95% уровню доверия.

```python
confidence_coef = 1.96
lower_catb = mean_catb_preds - confidence_coef * np.sqrt(var_catb_preds)
upper_catb = mean_catb_preds + confidence_coef * np.sqrt(var_catb_preds)
covered_catb = np.sum((y_test >= lower_catb) & (y_test <= upper_catb))
coverage_catb = covered_catb / len(y_test)
print(
    f'Процент предсказанных значений, попавший', 
    f'в интервал: {coverage_catb * 100: .2f} %.'
)
```

```python
eval_df_catb = pd.DataFrame({
    'Истинные значения': y_test,
    'Предсказанные средние': mean_catb_preds,
    'Нижняя граница': lower_catb,
    'Верхняя граница': upper_catb
})
eval_df_catb.reset_index(drop=True, inplace=True)
eval_df_catb[:5]
```

Посмотрим на графиках, как выглядят наши предсказания и границы 95% интервала.

```python
plt.figure(figsize=(15, 7))

sns.kdeplot(eval_df_catb['Истинные значения'],
            label='Истинные значения',
            fill=True,
            color='blue',
            alpha=0.7)
sns.kdeplot(eval_df_catb['Предсказанные средние'],
            label='Предсказанные средние',
            fill=True,
            color='red',
            alpha=0.5)
sns.kdeplot(eval_df_catb['Нижняя граница'],
            label='Нижняя граница',
            color='green',
            linestyle='--',
            alpha=0.7)
sns.kdeplot(eval_df_catb['Верхняя граница'],
            label='Верхняя граница',
            color='purple',
            linestyle='--',
            alpha=0.7)

plt.title(
    'Плотность истинных значений, предсказанных и границ интервала Catboost.')
plt.xlabel('Цена автомобиля')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

Теперь займёмся интервалами для LightGBM. Процесс будет немного отличаться от Catboost. В LightGBM мы будем предсказывать непосредственно квантили, в нашем случае 5% и 95%.

```python
lgbm_lower_model = LGBMRegressor(objective='quantile',
                                 alpha=0.05,
                                 **LGBM_PARAMS)
lgbm_upper_model = LGBMRegressor(objective='quantile',
                                 alpha=0.95,
                                 **LGBM_PARAMS)
lgbm_lower_model.fit(X_train_,
                     y_train_,
                     eval_set=[(X_val, y_val)],
                     eval_metric='mae')
lgbm_upper_model.fit(X_train_,
                     y_train_,
                     eval_set=[(X_val, y_val)],
                     eval_metric='mae')
lgbm_lower = lgbm_lower_model.predict(X_test)
lgbm_upper = lgbm_upper_model.predict(X_test)
covered_lgbm = np.sum((y_test >= lgbm_lower) & (y_test <= lgbm_upper))
coverage_lgbm = covered_lgbm / len(y_test)
print(
    f'Процент предсказанных значений, попавший', 
    f'в интервал: {coverage_lgbm * 100: .2f} %.'
)
```

```python
eval_df_lgbm = pd.DataFrame({
    'Истинные значения': y_test,
    'Предсказанные средние': lgbm_preds,
    'Нижняя граница': lgbm_lower,
    'Верхняя граница': lgbm_upper
})
eval_df_lgbm.reset_index(drop=True, inplace=True)
eval_df_lgbm[:5]
```

```python
plt.figure(figsize=(15, 7))

sns.kdeplot(eval_df_lgbm['Истинные значения'],
            label='Истинные значения',
            fill=True,
            color='blue',
            alpha=0.7)
sns.kdeplot(eval_df_lgbm['Предсказанные средние'],
            label='Предсказанные средние',
            fill=True,
            color='red',
            alpha=0.5)
sns.kdeplot(eval_df_lgbm['Нижняя граница'],
            label='Нижняя граница',
            color='green',
            linestyle='--',
            alpha=0.7)
sns.kdeplot(eval_df_lgbm['Верхняя граница'],
            label='Верхняя граница',
            color='purple',
            linestyle='--',
            alpha=0.7)

plt.title(
    'Плотность истинных значений, предсказанных и границ интервала LGBM.')
plt.xlabel('Цена автомобиля')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

LightGBM как то странно предсказал нижнюю границу, а также ошибается чуть больше, чем Catboost, но предсказательный интервал у LightGBM точнее на 26%! Также LightGBM обучился на порядок быстрее, чем Catboost.

# Feature importances.

Отлично! Теперь посмотрим на самые важные признаки, которые оказывают наибольшее влияние на предсказания модели.

```python
explainer_catb = shap.TreeExplainer(tuned_catb)
explanation_catb = explainer_catb(X_train)
shap.plots.beeswarm(explanation_catb[:, :, 0])
```

```python
explainer_lgbm = shap.TreeExplainer(tuned_lgbm)
explanation_lgbm = explainer_lgbm(X_train)
shap.plots.beeswarm(explanation_lgbm)
```

Ключевые признаки у моделей сильно схожи. Наиболее важные признаки - Пробег, Год_выпуска, Мощность_двигателя, Срок_эксплуатации. Это логично - чем, больше пробег, тем больше износ авто и как следствие ниже цена, что мы и наблюдаем на графике. Больший объём подразумевает больше мощность, что зачастую присуще более дорогим авто. Срок_эксплуатации и Год_выпуска связаны - чем меньше срок эксплуатации и больше год выпуска, тем больше цена. Мощность также связана с объёмом двигателя и также прямо коррелирует с ценой - чем больше тем дороже.

**Вывод:** Обе модели имеют свои плюсы и минусы. Подбор гиперпараметров через RandomSearch Catboost вообще не получилось применить - 3 итерации выполнялись почти 9 часов, а итоговый предсказательный интервал получился на 10% менее точным. LightGBM быстрее обучается и выдаёт более точный интервал. Вообще результативность моделей очень сильно менялась, в зависимости от размера исходного датасета, при первых экспериментах было около 27000 объектов, и лучше всех показывали себя RandomForest и XGB. Когда данных стало 32000 пальму первенства забрал LGBM. А при текущем размере в 37000 первое место делят Catboost и LightGBM. Если собрать еще больше данных, можно значительно улучшить качество.
